```python
import numpy as np
from matplotlib import pyplot as plt
import random
import scipy.stats as sps
import subprocess
```

```python
# записываем случайный граф с вероятностью ребра p
# в уже открытый файл file
def write_random_graph(vertex_count, p, file):
    for i in range(vertex_count - 1):
        for j in range(i + 1, vertex_count):
            frm = i
            to = j
            
            coin = sps.bernoulli.rvs(p=p, size=1)
            if coin == 1:
                file.write(str(frm) + ' ' + str(to) + '\n')
                
# записываем случайные веса вершин, распределенные
# равномерно на множестве {1, ..., n}, где n - число вершин в графе,
# в уже открытый файл file
def write_random_weights(vertex_count, file):
    for i in range(vertex_count):
        weight = 1 # random.randint(1, vertex_count)
        file.write(str(weight) + ' ')
    file.write('\n')
```

```python
def run_tests(vertex_count, graph_count, p, infile_name, outfile_name):
    is_correct_vertex_cover = np.array([])
    found_vertex_cover_weights = np.array([])
    min_vertex_cover_weights = np.array([])
    
    results = np.array([is_correct_vertex_cover, found_vertex_cover_weights,
                        min_vertex_cover_weights])
    
    for i in range(graph_count):
        infile = open(infile_name, 'w')
        
        infile.write(str(vertex_count) + '\n')
        write_random_weights(vertex_count, infile)
        write_random_graph(vertex_count, p, infile)
        infile.write('-1 -1' + '\n')
        
        infile.close()
        
        
        output = subprocess.check_output(['./cmake-build-debug/project']).decode('utf-8')
        
        
        outfile = open(outfile_name, 'r')
        
        result = np.array([])
        for line in outfile:
            index = line.find(":")
            result = np.append(result, int(line[(index + 2):]))
        results = np.append(results, [[result[0]], [result[1]], [result[2]]], axis=1)
        
        outfile.close()
        
    return results
```

```python
results = np.array([[], [], []])

infile_name = './tests_input/in.txt'
outfile_name = './tests_output/out.txt'

vertex_count = 3
graph_count = 2
p = 1
result = run_tests(vertex_count, graph_count, p, infile_name, outfile_name)
results = np.append(results, result, axis=1)

vertex_count = 4
graph_count = 10
p = 1
result = run_tests(vertex_count, graph_count, p, infile_name, outfile_name)
results = np.append(results, result, axis=1)

vertex_count = 5
graph_count = 100
p = 1
result = run_tests(vertex_count, graph_count, p, infile_name, outfile_name)
results = np.append(results, result, axis=1)

vertex_count = 6
graph_count = 200
p = 1
result = run_tests(vertex_count, graph_count, p, infile_name, outfile_name)
results = np.append(results, result, axis=1)

vertex_count = 7
graph_count = 300
p = 1
result = run_tests(vertex_count, graph_count, p, infile_name, outfile_name)
results = np.append(results, result, axis=1)

vertex_count = 8
graph_count = 400
p = 1
result = run_tests(vertex_count, graph_count, p, infile_name, outfile_name)
results = np.append(results, result, axis=1)
```
